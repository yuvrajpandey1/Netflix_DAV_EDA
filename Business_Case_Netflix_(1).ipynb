### Top 5 Countries Where Netflix is Popular


```python
df['split_country'] = df['split_country'].str.rstrip()
df
```

```python
country_counts = df['split_country'].value_counts()
country_counts
```

```python
country_counts = df['split_country'].value_counts().sum()
country_counts
```

```python
top_5_countries = country_counts.head(5)
```

```python
plt.figure(figsize=(16, 10))
bar_plot = sns.barplot(x=top_5_countries.index, y=top_5_countries.values)
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.title('Top 5 Countries Where Netflix is Popular')
```
